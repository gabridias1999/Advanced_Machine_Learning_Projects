```python
df["Time"] = None
for ind, row in df.iterrows():
    new = False
    if ind >= 40:
        new = True
    start_time = time.time()
    print(ind)
    if row["Type"] == "RF":
        params = row["Params"]
        if new:
            RF(row["Dataset"], params["criterion"], params["max_depth"], params["n_estimators"], True)
        else:
            RF(row["Dataset"], params["criterion"], params["max_depth"], None, False)

    elif row["Type"] == "SVM":
        params = row["Params"]
        if new:
            pass
        else:
            SVM(row["Dataset"], params["kernel"])
        

    elif row["Type"] == "LR":
        params = row["Params"]
        if new:
            LR(row["Dataset"], params["penalty"], params["solver"], True)
        else:
            LR(row["Dataset"], params["penalty"], None, False)
    
    elif row["Type"] == "KNN":
        params = row["Params"]
        if new:
            KNN(row["Dataset"], params["n_neighbors"], params["weights"], params["p"], True)
        else:
            KNN(row["Dataset"], params["n_neighbors"], params["weights"], None, False)
    
    elif row["Type"] == "Boosting":
        params = row["Params"]
        if new:
            pass
        else:
            Boosting(row["Dataset"], params["learning_rate"])

    end_time = time.time()
    df.at[ind, "Time"] = int(end_time - start_time)
    #display(df)
    df.to_pickle("/drive/MyDrive/ACA/Projeto/results_TVT.pkl")
    df.to_csv("/drive/MyDrive/ACA/Projeto/results_TVT.csv")

```
