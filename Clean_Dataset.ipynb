```python
def check_outliers(coluna, df):
    try:
        q1 = df[coluna].quantile(0.25)
        q3 = df[coluna].quantile(0.75)
        IQR = q3-q1

        outliers = df[coluna][((df[coluna]<(q1-1.5*IQR)) | (df[coluna]>(q3+1.5*IQR)))]
        return outliers
    except TypeError:
        return []

def clean_train(df_train, change_outliers, change_corr, change_norm, change_pca, num_components):
    df = df_train.copy()
    filename = ""

    #========================================================================
    # Substituir valores categoricos por numericos
    print("Substituir valores categoricos por numericos..")
    df["attack_type"] = df["attack_type"].map({"normal": 0, "Dos": 1, "R2L": 2, "U2R": 3, "Probe": 4})
    df["protocol_type"] = df["protocol_type"].map({'tcp': 0, 'udp': 1, 'icmp': 2})
    df["flag"] = df["flag"].map({'SF': 0, 'S0': 1, 'RSTO': 2, 'REJ': 3, 'SH': 4, 'RSTR': 5, 'S2': 6, 'S3': 7, 'S1': 8, 'RSTOS0': 9, 'OTH': 10})
    df["service"] = df["service"].map({'smtp': 0, 'login': 1, 'http': 2, 'private': 3, 'telnet': 4, 'other': 5, 'ecr_i': 6, 'iso_tsap': 7, 'csnet_ns': 8, 'ftp_data': 9, 'imap4': 10, 'ftp': 11, 'time': 12, 'daytime': 13, 'pop_3': 14, 'kshell': 15, 'uucp': 16, 'name': 17, 'domain_u': 18, 'echo': 19, 'whois': 20, 'eco_i': 21, 'rje': 22, 'finger': 23, 'Z39_50': 24, 'courier': 25, 'discard': 26, 'domain': 27, 'auth': 28, 'supdup': 29, 'netbios_ns': 30, 'IRC': 31, 'netstat': 32, 'nnsp': 33, 'netbios_dgm': 34, 'link': 35, 'hostnames': 36, 'klogin': 37, 'http_443': 38, 'systat': 39, 'uucp_path': 40, 'netbios_ssn': 41, 'exec': 42, 'sunrpc': 43, 'bgp': 44, 'nntp': 45, 'mtp': 46, 'ssh': 47, 'ntp_u': 48, 'ldap': 49, 'urp_i': 50, 'gopher': 51, 'ctf': 52, 'efs': 53, 'sql_net': 54, 'tim_i': 55, 'vmnet': 56, 'urh_i': 57, 'shell': 58, 'remote_job': 59, 'X11': 60, 'tftp_u': 61, 'pop_2': 62, 'red_i': 63, 'printer': 64, 'pm_dump': 65, 'aol': 66, 'http_8001': 67, 'harvest': 68})
    
    #========================================================================
    # Verificar outliers -> Interquartile Range (IQR) Method
    print("Verificar outliers -> Interquartile Range (IQR) Method..")
    params_out_col = []
    params_y_q1 = []
    params_y_q3 = []
    if change_outliers:
        filename += "NoOutliers_"
        for col in df.columns:
            if col != "attack_type":
                outliers = check_outliers(col, df)

                if len(outliers) > 0:
                    if len(df[col].unique()) > 10:
                        #print(len(df), end= " -> ")
                        
                        q1 = df[col].quantile(0.25)
                        q3 = df[col].quantile(0.75)
                        IQR = q3-q1
                        y_q1 =  (q1 - 1.5 * IQR)
                        y_q3 =  (q3 + 1.5 * IQR)

                        params_out_col.append(col)
                        params_y_q1.append(y_q1)
                        params_y_q3.append(y_q3)


                        for ind in outliers.index:
                            val = outliers[ind]
                            if val <= y_q1:
                                df.at[ind, col] = y_q1
                            elif val >= y_q3:
                                df.at[ind, col] = y_q3
                        
                        #print(len(df))
                        df = df.reset_index(drop= True)
        
    else:
        filename += "Outliers_"

    #========================================================================
    # Verificar unique values
    # Procuramos colunas com apenas um valor
    print("Verificar unique values..")
    print("DF -> ", len(df.columns), end= " -> ")
    colunas = []
    
    for col in df.columns:
        if col != "attack_type":
            if len(df[col].unique()) == 1:
                colunas.append(col)

    df = df.drop(columns = colunas )
    print(len(df.columns))

    params_drop_unique = colunas

    #========================================================================
    # Temos muitas colunas e algumas delas sem correlação com o nosso target.
    print("Ver correlation..")
    params_drop_corr = []
    if change_corr:
        filename += "Corr_"
        for col in df.columns:
            try:
                corr = df["attack_type"].corr( df[str(col)] )
                if abs(corr) <= 0.01:
                    #print(col , " -> ", corr)
                    df = df.drop(columns = col )
                    params_drop_corr.append(col)
            except TypeError:
                pass
    else:
        filename += "NoCorr_"
    

    #========================================================================
    # Normalization ou Standardization
    params_norm_vals = []
    print("Normalization..")
    if change_norm:
        filename += "ST_"
        colunas_normalizar = ["duration", "src_bytes", "dst_bytes", "hot", "num_compromised", "num_root", "num_file_creations", 
                                "count", "srv_count", "dst_host_count", "dst_host_srv_count"]

        for col in colunas_normalizar:
            if col in df.columns:
                if change_outliers: # normalization
                    params_norm_vals.append([df[col].min(), df[col].max()])
                    df[col] = (df[col] - df[col].min()) / (df[col].max()- df[col].min())
                else: # standardization
                    params_norm_vals.append([df[col].mean(), df[col].std()])
                    df[col] = (df[col] - df[col].mean()) / df[col].std()
    else:
        filename += "NoST_"
    print("Shape: ", np.shape(df))
    
    #========================================================================
    # Guarda media da col
    params_mean = []
    params_mean_col = []
    for col in df.columns:
        params_mean_col.append(col)
        params_mean.append(df[col].mean())

    #========================================================================
    # PCA
    print("PCA..")
    params_pca = 0
    pca_train = None
    if change_pca:
        params_pca = num_components
        filename += "PCA"
       
        pca = PCA(n_components= num_components)
        colunas = df.columns.to_list()
        colunas.remove("attack_type")

        pca.fit(df[colunas])
        pca_train = pca
        print("PCA ratio: ", pca.explained_variance_ratio_)
        print("PCA info: ", np.sum(pca.explained_variance_ratio_))
        #print(pca.singular_values_)

        df_pca = pca.transform(df[colunas])

        print(np.shape(df_pca))

        df_dict = {}
        for i in range(num_components):
            df_dict["PCA_" + str(i+1)] = df_pca[:,i]
        df_dict["Target"] = df["attack_type"]

        df = pd.DataFrame( df_dict)
    else:
       filename += "NoPCA" 

    print("Filename: ", filename)
    df.to_csv("/drive/MyDrive/ACA/Projeto/dataset/Clean/Train/" + filename + ".csv", index = False)

    # Fazer o mesmo para o test dataset
    params_test.loc[filename] = [params_out_col, params_y_q1, params_y_q3, params_mean, params_mean_col, params_drop_unique, params_drop_corr, params_norm_vals, params_pca ]
    display(params_test)
    clean_test(change_outliers, change_corr, change_norm, change_pca, filename, pca_train)
```
